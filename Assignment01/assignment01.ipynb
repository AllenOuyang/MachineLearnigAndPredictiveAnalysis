```python
import pandas as pd
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
```

```python
# define the path of data
data_path = "/Users/allenouyang/jupyter/2023Sprint ML&PA(MSU)/Assignment01"
data = pd.read_csv(os.path.join(data_path, "test_sample.csv"))
```

```python
data.head()
```

```python
# Split data into X and Y
X = data.iloc[:,1:]
Y = data.iloc[:,0]

# Initialize variables
r2 = []
n_orig = None

# Fit linear regression models with increasing number of regressors
for i in range(1, len(X.columns)+1):
    lin_reg = LinearRegression()
    lin_reg.fit(X.iloc[:,:i], Y)
    r2.append(lin_reg.score(X.iloc[:,:i], Y))
    if n_orig is None and r2[-1] > 0.9:
        n_orig = i
        
# Print results
print("Smallest number of regressors making determination coefficient greater than 0.9: ", n_orig)

```

```python
# Fit PCA on the data
pca = PCA()
X_PCA = pca.fit_transform(X)

# Calculate explained variance ratio for each component
variance_ratio = pca.explained_variance_ratio_

# Calculate cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(variance_ratio)
# Initialize variables
n_PCA = None
r2_PCA = []

# Find smallest number of PCA factors necessary for determination coefficient greater than 0.9
for i in range(1, n_orig+1):
    lin_reg = LinearRegression()
    lin_reg.fit(X_PCA[:,:i], Y)
    r2_PCA.append(lin_reg.score(X_PCA[:,:i], Y))
    if n_PCA is None and r2_PCA[-1] > 0.9:
        n_PCA = i
        break
        
# Print results
print("Smallest number of PCA factors necessary for determination coefficient greater than 0.9: ", n_PCA)
print("Determination coefficient for the model with n_PCA selected most important meta-features: ", r2_PCA[-1])
```
